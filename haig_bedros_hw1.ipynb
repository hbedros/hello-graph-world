For this assignment we are using Networkx package for Python.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a new graph
graph = nx.Graph()

# Add nodes (names)
nodes = ["Andre", "Beverly", "Carol", "Diane", "Ed", "Garth", "Fernando", "Heather", "Ike", "Jane", "Garth"]
graph.add_nodes_from(nodes)

# Add edges
edges = [
    ("Andre", "Carol"),
    ("Andre", "Fernando"),
    ("Andre", "Diane"),
    ("Andre", "Beverly"),
    ("Beverly", "Andre"),
    ("Beverly", "Diane"),
    ("Beverly", "Garth"),
    ("Beverly", "Ed"),
    ("Carol", "Fernando"),
    ("Carol", "Diane"),
    ("Diane", "Fernando"),
    ("Diane", "Garth"),
    ("Diane", "Ed"),
    ("Ed", "Garth"),
    ("Fernando", "Heather"),
    ("Garth", "Heather"),
    ("Heather", "Ike"),
    ("Ike", "Jane"),
       
]
graph.add_edges_from(edges)

# Plot the graph
plt.figure(figsize=(10, 8))
nx.draw(graph, with_labels=True, node_color='lightgrey', node_size=3000, font_size=12, font_color='black', edge_color='purple')
plt.show()

```
